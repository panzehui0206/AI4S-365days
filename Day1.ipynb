**Biopython**是一套专门为计算生物学和生物信息学设计的 **Python** 工具包。它的核心使命是：让计算机能够像生物学家一样“读懂”和“操作”生命分子。

**from Bio.Seq import Seq** 引入了 **Biopython** 的核心序列对象，它赋予了 DNA 字符串“生物学超能力”。


```python
# =============================================
# Cell 2：Bio.Seq的功能
# =============================================
from Bio.Seq import Seq

# 创建一条 DNA 序列
dna = Seq("ATGGCCATTGTAATGGGCCGCTGAAAGGGTGCCCGATAG")

# 1. 转录：DNA → mRNA
mrna = dna.transcribe()
print(mrna)  # AUGGCCAUUGUAAUGGGCCGCUGAAAGGGUGCCCGAUAG

# 2. 反向互补（获取互补链的 5'→3' 方向序列）
rc_dna = dna.reverse_complement()
print(rc_dna)  # CTATCGGGCACCCTTTCAGCGGCCCATTACAATGGCCAT

# 3. 翻译：mRNA → 蛋白质（使用标准遗传密码表）
protein = dna.translate()  # 也可以对 mRNA 调用，效果相同
print(protein)  # MAIVMGR*KGAR*   （* 代表终止密码子）

# 4. 翻译时指定不同的密码表（如线粒体遗传密码）
protein_mito = dna.translate(table=2)  # 2 代表脊椎动物线粒体密码表
print(protein_mito)

# 5. 切片与拼接（和字符串用法一致）
dna[:10]                # 前10个碱基
dna[1:10:3]             # 步长切片
dna + Seq("TTTT")       # 拼接两条序列
```

这就是**Bio.Seq**，用来处理模型输入输出的 **DNA/RNA/蛋白序列**，有序列对象、转录、翻译、反向互补等功能。

```python
# =============================================
# Cell 3：人类胰岛素基因编码区
# =============================================
insulin_dna = Seq("ATGGCCCTGTGGATGCGCCTCCTGCCCCTGCTGGCGCTGCTGGCCCTCTGGGGACCTGACCCAGCCGCAGCCTTTGTGAACCAACACCTGTGCGGCTCACACCTGGTGGAAGCTCTCTACCTAGTGTGCGGGGAACGAGGCTTCTTCTACACACCCAAGACCCGCCGGGAGGCAGAGGACCTGCAGGTGGGGCAGGTGGAGCTGGGCGGGGGCCCTGGTGCAGGCAGCCTGCAGCCCTTGGCCCTGGAGGGGTCCCTGCAGAAGCGTGGCATTGTGGAACAATGCTGTACCAGCATCTGCTCCCTCTACCAGCTGGAGAACTACTGCAACTAG")

print(f"📜 原始 DNA 序列 (前50个碱基): {insulin_dna[:50]}...")
print(f"📏 序列长度: {len(insulin_dna)} bp")
```

我们用 **Seq**() 把一串字符串变成了一个 **Biopython** 序列对象。

这意味着它不再只是普通的文本，而是一条有生物学属性的 **DNA** 链。

这条序列是人类胰岛素基因的**真实编码区**（**CDS**）。

**编码区**（**CDS**，全称 **Coding DNA Sequence**）是指 **DNA** 或 **RNA** 分子中真正负责编码蛋白质氨基酸序列的那一段区域。

区分编码区和完整基因非常重要，因为在不同的任务中，需要用到不同的序列类型。

**完整基因** ≈ **编码区** + **内含子** + **非翻译区（UTR）** + **调控区**（长达数千到数百万碱基）。

**编码区** = 从**起始密码子** 到**终止密码子**的连续片段（通常只有几百到几千碱基）。



```python
# =============================================
# Cell 4：转录（DNA -> mRNA）
# =============================================
insulin_mrna = insulin_dna.transcribe()
print(f"🧬 转录后的 mRNA (前50个碱基): {insulin_mrna[:50]}...")
```

转录是以 **DNA** 为模板合成 **RNA** 的过程，将遗传信息从 **DNA** 复制到 **mRNA** 上，在细胞内，这是由 **RNA** 聚合酶完成的。

在代码中，**.transcribe**() 方法自动将 **DNA** 序列中的 **T** (胸腺嘧啶) 替换为 **U** (尿嘧啶)。




```python
# =============================================
# Cell 5：翻译（mRNA -> 蛋白质）
# =============================================
insulin_protein = insulin_mrna.translate()
print(f"🍳 翻译后的蛋白质序列:\n{insulin_protein}")
```

**.translate**() 方法从第一个碱基开始，每三个一组（**密码子**）查表翻译成氨基酸。

序列末尾的 * 就是终止密码子，代表蛋白质合成到此为止。

**MALWMR...** 就是胰岛素蛋白的氨基酸序列（单字母缩写）。

```python
# =============================================
# Cell 6：计算 GC 含量
# =============================================
gc_count = insulin_dna.count('G') + insulin_dna.count('C')
gc_percent = 100 * gc_count / len(insulin_dna)
print(f"📊 GC 含量: {gc_percent:.2f}%")
print("💡 GC 含量越高，DNA 双链越稳定（G≡C 之间有三个氢键）")
```

**GC** 含量是分子生物学实验（如 **PCR**）的重要参数，因为它影响 **DNA** 双链的解链温度（**Tm**）。

这里用 **Python** 自带的 .**count**() 方法统计 **G** 和 **C** 的数量，简单高效。

```python
# =============================================
# Cell 7：定位终止密码子
# =============================================
stop_positions = [i for i, aa in enumerate(insulin_protein) if aa == '*']
if stop_positions:
    print(f"🛑 终止密码子出现在位置: {stop_positions}")
else:
    print("⚠️ 未找到终止密码子，序列可能不完整")
```

**enumerate**() 同时给出索引和氨基酸字母。

列表推导式 [i for i, aa in ... if aa == '*'] 收集所有 * 的位置。

正常情况下，真核生物基因在末尾只有一个终止密码子。
